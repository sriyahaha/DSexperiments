```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df[numeric_columns] < (Q1 - 1.5 * IQR)) | (df[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]
df
```

Line charts with legends and markers
Line Chart
A line chart is a visual representation of data points connected by straight line segments. It's primarily used to show trends or changes over time.

Horizontal axis: Usually represents time (days, months, years).
Vertical axis: Represents the values of the data being measured.
Key features:

Clarity: Easily shows trends, patterns, and fluctuations.
Comparison: Can compare multiple data sets simultaneously.
Predictions: Helps in forecasting future values based on existing trends.
Common uses:

Stock prices over time
Temperature changes
Website traffic
Sales figures

```python
import pandas as pd
import matplotlib.pyplot as plt
# Line chart with legends and markers
plt.figure(figsize=(10, 6))

# Plotting the first line with markers
plt.plot(data['dteday'], data['season'], marker='o', linestyle='-', color='blue', label='dteday')

# Plotting the second line with different markers
plt.plot(data['dteday'], data['season'], marker='s', linestyle='--', color='red', label='season')

# Adding title and labels
plt.title('Line Chart with Legends and Markers', fontsize=16)
plt.xlabel('dteday', fontsize=14)
plt.ylabel('season', fontsize=14)

# Adding a legend
plt.legend()

# Display the plot
plt.show()
```


Count Plot
A count plot is a visual representation of the count of observations in each categorical bin. It's essentially a bar chart for categorical data.

Horizontal or vertical bars: Represent the different categories.
Height or length of the bars: Indicates the frequency or count of each category.
Key features:

Distribution: Shows the distribution of categorical data.
Comparison: Allows easy comparison of category frequencies.
Outliers: Can identify categories with significantly higher or lower counts.
Common uses:

Frequency of different product categories
Distribution of customer demographics (gender, age group)
Count of different outcomes in a survey

Count Plot Using Matplotlib

```python
# Count the occurrences of each category
category_counts = data['mnth'].value_counts()

# Plot the counts
plt.figure(figsize=(10, 6))
category_counts.plot(kind='bar', color='skyblue')

# Adding title and labels
plt.title('Count Plot', fontsize=16)
plt.xlabel('mnth', fontsize=14)
plt.ylabel('yr', fontsize=14)

# Display the plot
plt.show()
```

Histograms - A histogram is a type of data visualization used to represent the distribution of a dataset. It displays the frequency of data points within certain ranges or bins. Each bar in a histogram represents a bin or range of values, with the height of the bar indicating the number of data points that fall within that range. Histograms are useful for understanding the shape of data distribution, detecting patterns, and identifying outliers.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/bike.csv')

# Display the first few rows of the dataset to understand its structure
print(data.head())

# Create a histogram for the 'weathersit' column
plt.figure(figsize=(10, 6))
plt.hist(data['weathersit'], bins=20, color='skyblue', edgecolor='black')

# Adding titles and labels
plt.title('Histogram of Weather Situation', fontsize=16)
plt.xlabel('Weather Situation', fontsize=14)
plt.ylabel('Frequency', fontsize=14)

# Display the plot
plt.show()
```

Adding Hues - Hues in histograms refer to the use of different colors to represent different categories or groups within the dataset. Adding hues allows you to visualize how the distribution of data varies across these categories, providing more depth and clarity to the analysis.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/bike.csv')

# Display the first few rows of the dataset to understand its structure
print(data.head())

# Create a histogram with hues
plt.figure(figsize=(12, 8))

# List unique categories in 'weathersit'
weather_categories = data['weathersit'].unique()

# Assign colors for each category
colors = plt.get_cmap('tab10', len(weather_categories))  # 'tab10' colormap provides distinct colors

# Plot histogram for each category
for category in weather_categories:
    subset = data[data['weathersit'] == category]
    plt.hist(subset['weathersit'], bins=20, alpha=0.7, label=f'Weather Situation {category}', color=colors(weather_categories.tolist().index(category)))

# Adding titles and labels
plt.title('Histogram of Weather Situation by Category', fontsize=16)
plt.xlabel('Weather Situation', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.legend(title='Weather Situation')

# Display the plot
plt.show()
```
