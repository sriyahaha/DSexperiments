```python
df.drop_duplicates(inplace=True)
```

```python
df
```

```python
features = df.columns.tolist()
plt.figure(figsize=(20,15))
for i, feature in enumerate(features):
    plt.subplot(6, 4, i + 1)
    sns.boxplot(data=df[feature])
    plt.title(f'Box plot of {feature}')
plt.tight_layout()
plt.show()
```

Using IQR to handle Outliers

```python
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns
Q1 = df[numeric_columns].quantile(0.25)
Q3 = df[numeric_columns].quantile(0.75)
IQR = Q3 - Q1
df = df[~((df[numeric_columns] < (Q1 - 1.5 * IQR)) | (df[numeric_columns] > (Q3 + 1.5 * IQR))).any(axis=1)]
df
```

```python
features = df.columns.tolist()
# Plot box plots for each feature
plt.figure(figsize=(20, 15))
for i, feature in enumerate(features):
    plt.subplot(6, 4, i + 1)
    sns.boxplot(data=df[feature])
    plt.title(f'Box plot of {feature}')
plt.tight_layout()
plt.show()
```

```python
df=pd.read_csv('/content/bank.csv')
df
```

Encoding Categorical Variables

```python
print(df.columns)
```

One Hot Encoding

```python
df = pd.get_dummies(df, columns=['housing'])

df
```

Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['deposit']=le.fit_transform(df['deposit'])
df
```

Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
size_mapping = {'tertiary':0, 'secondary':1, 'unknown':2}
df['education']=df['education'].map(size_mapping)
df
```

Frequency Encoding

```python
frequency_encoding = df['job'].value_counts().to_dict()
df['job'] =df['job'].map(frequency_encoding)
df
```

Target Encoding

```python
mean_encoded = df.groupby('balance')['loan'].mean().to_dict()
df['balance'] = df['balance'].map(mean_encoded)
df
```

Hashing Encoding

```python
df['marital'] = df['marital'].astype(str)
df
```

```python
import pandas as pd
from sklearn.feature_extraction import FeatureHasher

# Sample DataFrame with potential missing values
data = {
    'marital': ['single', 'married', 'divorced', 'married', 'single', None],
    'loan': [1, 0, 1, 0, 0, 1]
}

df = pd.DataFrame(data)
df

```

```python
marital_list = df['marital'].apply(lambda x: [x]).tolist()
df
```

```python
hasher = FeatureHasher(input_type='string', n_features=8)
hashed_features = hasher.transform(marital_list)
df_hashed = pd.DataFrame(hashed_features.toarray(), columns=[f'marital_{i}' for i in range(8)])

print(df_hashed)
```
